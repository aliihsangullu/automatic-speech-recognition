```python
#---------importing libraries--------
import pyaudio
import wave
from python_speech_features import mfcc
import sys
import numpy as np
from scipy.signal import butter, lfilter, freqz
import matplotlib.pyplot as plt
np.set_printoptions(threshold=sys.maxsize)
import scipy.io.wavfile as wav
import math
from fastdtw import fastdtw
#-------------------------------------

#---------LowPassFilter-------------
def butter_lowpass(cutoff, fs, order):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def butter_lowpass_filter(data, cutoff, fs, order):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = lfilter(b, a, data)
    return y

# Filter requirements.
order = 7
fs = 16000      # sample rate, Hz
cutoff = 250  # desired cutoff frequency of the filter, Hz

#-----------------------------------------

CHUNK = 16000
FORMAT = pyaudio.paInt32
CHANNELS = 1
RATE = 16000 #sampling
RECORD_SECONDS = 4 #record time

#--------------Three times voice record--------------
for clock in range(0,3):

    WAVE_OUTPUT_FILENAME = "database/output"+str(clock)+".wav"# creation of wav file

    p = pyaudio.PyAudio()

    stream = p.open(format=FORMAT,
                channels=CHANNELS,
                rate=RATE,
                input=True,
                frames_per_buffer=CHUNK)

    print("* recording")

    frames = []

    for i in range(0, int(RATE / CHUNK * RECORD_SECONDS)):
        data = stream.read(CHUNK)
        frames.append(data)
    print("* done recording")

    stream.stop_stream()
    stream.close()
    p.terminate()

    wf = wave.open(WAVE_OUTPUT_FILENAME, 'wb')
    wf.setnchannels(CHANNELS)
    wf.setsampwidth(p.get_sample_size(FORMAT))
    wf.setframerate(RATE)
    wf.writeframes(b''.join(frames))
    wf.close()
    
print("Your voice has recorded succesfully.")
#---------------------------------------------------------

result = []#for print


for clock in range(0,3):
    (RATE, sig) = wav.read("database/output"+str(clock)+".wav")# read to created files
    mfcc_output = mfcc(sig, RATE) #Mel-frequency cepstrum
    mfcc_outputFilter = butter_lowpass_filter(mfcc_output, cutoff, fs, order)# into LPF to LPF Mel-frequency cepstrum


    #---------database samples----------
    #reading from database()
    #getting Mel-frequency cepstrum of read files

    (RATE, sig) = wav.read("database/a1.wav")
    mfcc_a1 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/b1.wav")
    mfcc_b1 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/c1.wav")
    mfcc_c1 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/d1.wav")
    mfcc_d1 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/e1.wav")
    mfcc_e1 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/f1.wav")
    mfcc_f1 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/g1.wav")
    mfcc_g1 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/h1.wav")
    mfcc_h1 = mfcc(sig, RATE)

        #--------------1.part end--------------

    (RATE, sig) = wav.read("database/a2.wav")
    mfcc_a2 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/b2.wav")
    mfcc_b2 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/c2.wav")
    mfcc_c2 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/d2.wav")
    mfcc_d2 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/e2.wav")
    mfcc_e2 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/f2.wav")
    mfcc_f2 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/g2.wav")
    mfcc_g2 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/h2.wav")
    mfcc_h2 = mfcc(sig, RATE)

        #------------2.part end-----------------

    (RATE, sig) = wav.read("database/a3.wav")
    mfcc_a3 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/b3.wav")
    mfcc_b3 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/c3.wav")
    mfcc_c3 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/d3.wav")
    mfcc_d3 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/e3.wav")
    mfcc_e3 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/f3.wav")
    mfcc_f3 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/g3.wav")
    mfcc_g3 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/h3.wav")
    mfcc_h3 = mfcc(sig, RATE)

        #-------------3.part end-------------

    (RATE, sig) = wav.read("database/a4.wav")
    mfcc_a4 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/b4.wav")
    mfcc_b4 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/c4.wav")
    mfcc_c4 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/d4.wav")
    mfcc_d4 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/e4.wav")
    mfcc_e4 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/f4.wav")
    mfcc_f4 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/g4.wav")
    mfcc_g4 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/h4.wav")
    mfcc_h4 = mfcc(sig, RATE)

        #------------4.part end------------------

    (RATE, sig) = wav.read("database/a5.wav")
    mfcc_a5 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/b5.wav")
    mfcc_b5 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/c5.wav")
    mfcc_c5 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/d5.wav")
    mfcc_d5 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/e5.wav")
    mfcc_e5 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/f5.wav")
    mfcc_f5 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/g5.wav")
    mfcc_g5 = mfcc(sig, RATE)

    (RATE, sig) = wav.read("database/h5.wav")
    mfcc_h5 = mfcc(sig, RATE)

    #------------5.part end--------------
    #----------database End--------------


    #Dynamic time warping  
    a1=fastdtw(mfcc_outputFilter, mfcc_a1)
    b1=fastdtw(mfcc_outputFilter, mfcc_b1)
    c1=fastdtw(mfcc_outputFilter, mfcc_c1)
    d1=fastdtw(mfcc_outputFilter, mfcc_d1)
    e1=fastdtw(mfcc_outputFilter, mfcc_e1)
    f1=fastdtw(mfcc_outputFilter, mfcc_f1)
    g1=fastdtw(mfcc_outputFilter, mfcc_g1)
    h1=fastdtw(mfcc_outputFilter, mfcc_h1)
    #--------1.part end---------

    a2=fastdtw(mfcc_outputFilter, mfcc_a2)
    b2=fastdtw(mfcc_outputFilter, mfcc_b2)
    c2=fastdtw(mfcc_outputFilter, mfcc_c2)
    d2=fastdtw(mfcc_outputFilter, mfcc_d2)
    e2=fastdtw(mfcc_outputFilter, mfcc_e2)
    f2=fastdtw(mfcc_outputFilter, mfcc_f2)
    g2=fastdtw(mfcc_outputFilter, mfcc_g2)
    h2=fastdtw(mfcc_outputFilter, mfcc_h2)
    #----------2.part end-------

    a3=fastdtw(mfcc_outputFilter, mfcc_a3)
    b3=fastdtw(mfcc_outputFilter, mfcc_b3)
    c3=fastdtw(mfcc_outputFilter, mfcc_c3)
    d3=fastdtw(mfcc_outputFilter, mfcc_d3)
    e3=fastdtw(mfcc_outputFilter, mfcc_e3)
    f3=fastdtw(mfcc_outputFilter, mfcc_f3)
    g3=fastdtw(mfcc_outputFilter, mfcc_g3)
    h3=fastdtw(mfcc_outputFilter, mfcc_h3)
    #----------3.part end--------

    a4=fastdtw(mfcc_outputFilter, mfcc_a4)
    b4=fastdtw(mfcc_outputFilter, mfcc_b4)
    c4=fastdtw(mfcc_outputFilter, mfcc_c4)
    d4=fastdtw(mfcc_outputFilter, mfcc_d4)
    e4=fastdtw(mfcc_outputFilter, mfcc_e4)
    f4=fastdtw(mfcc_outputFilter, mfcc_e4)
    g4=fastdtw(mfcc_outputFilter, mfcc_g4)
    h4=fastdtw(mfcc_outputFilter, mfcc_h4)
    #----------4.part end-------

    a5=fastdtw(mfcc_outputFilter, mfcc_a5)
    b5=fastdtw(mfcc_outputFilter, mfcc_b5)
    c5=fastdtw(mfcc_outputFilter, mfcc_c5)
    d5=fastdtw(mfcc_outputFilter, mfcc_d5)
    e5=fastdtw(mfcc_outputFilter, mfcc_e5)
    f5=fastdtw(mfcc_outputFilter, mfcc_f5)
    g5=fastdtw(mfcc_outputFilter, mfcc_g5)
    h5=fastdtw(mfcc_outputFilter, mfcc_h5)
    #----------5.part end--------

    a1=fastdtw(mfcc_output, mfcc_a1)
    b1=fastdtw(mfcc_output, mfcc_b1)
    c1=fastdtw(mfcc_output, mfcc_c1)
    d1=fastdtw(mfcc_output, mfcc_d1)
    e1=fastdtw(mfcc_output, mfcc_e1)
    f1=fastdtw(mfcc_output, mfcc_f1)
    g1=fastdtw(mfcc_output, mfcc_g1)
    h1=fastdtw(mfcc_output, mfcc_h1)
    #--------1.part end---------

    a2=fastdtw(mfcc_output, mfcc_a2)
    b2=fastdtw(mfcc_output, mfcc_b2)
    c2=fastdtw(mfcc_output, mfcc_c2)
    d2=fastdtw(mfcc_output, mfcc_d2)
    e2=fastdtw(mfcc_output, mfcc_e2)
    f2=fastdtw(mfcc_output, mfcc_f2)
    g2=fastdtw(mfcc_output, mfcc_g2)
    h2=fastdtw(mfcc_output, mfcc_h2)
    #----------2.part end-------

    a3=fastdtw(mfcc_output, mfcc_a3)
    b3=fastdtw(mfcc_output, mfcc_b3)
    c3=fastdtw(mfcc_output, mfcc_c3)
    d3=fastdtw(mfcc_output, mfcc_d3)
    e3=fastdtw(mfcc_output, mfcc_e3)
    f3=fastdtw(mfcc_output, mfcc_f3)
    g3=fastdtw(mfcc_output, mfcc_g3)
    h3=fastdtw(mfcc_output, mfcc_h3)
    #----------3.part end--------

    a4=fastdtw(mfcc_output, mfcc_a4)
    b4=fastdtw(mfcc_output, mfcc_b4)
    c4=fastdtw(mfcc_output, mfcc_c4)
    d4=fastdtw(mfcc_output, mfcc_d4)
    e4=fastdtw(mfcc_output, mfcc_e4)
    f4=fastdtw(mfcc_output, mfcc_e4)
    g4=fastdtw(mfcc_output, mfcc_g4)
    h4=fastdtw(mfcc_output, mfcc_h4)
    #----------4.part end-------

    a5=fastdtw(mfcc_output, mfcc_a5)
    b5=fastdtw(mfcc_output, mfcc_b5)
    c5=fastdtw(mfcc_output, mfcc_c5)
    d5=fastdtw(mfcc_output, mfcc_d5)
    e5=fastdtw(mfcc_output, mfcc_e5)
    f5=fastdtw(mfcc_output, mfcc_f5)
    g5=fastdtw(mfcc_output, mfcc_g5)
    h5=fastdtw(mfcc_output, mfcc_h5)
    #----------5.part end--------

    dtwlist=[]
    
    

    dtwlist.append(a1)
    dtwlist.append(b1)
    dtwlist.append(c1)
    dtwlist.append(d1)
    dtwlist.append(e1)
    dtwlist.append(f1)
    dtwlist.append(h1)
    dtwlist.append(g1)
    #----------1.part end-------

    dtwlist.append(a2)
    dtwlist.append(b2)
    dtwlist.append(c2)
    dtwlist.append(d2)
    dtwlist.append(e2)
    dtwlist.append(f2)
    dtwlist.append(h2)
    dtwlist.append(g2)
    #--------2.part end--------

    dtwlist.append(a3)
    dtwlist.append(b3)
    dtwlist.append(c3)
    dtwlist.append(d3)
    dtwlist.append(e3)
    dtwlist.append(f3)
    dtwlist.append(h3)
    dtwlist.append(g3)
    #----------3.part end---------

    dtwlist.append(a4)
    dtwlist.append(b4)
    dtwlist.append(c4)
    dtwlist.append(d4)
    dtwlist.append(e4)
    dtwlist.append(f4)
    dtwlist.append(h4)
    dtwlist.append(g4)
    #--------------4.part end--------

    dtwlist.append(a5)
    dtwlist.append(b5)
    dtwlist.append(c5)
    dtwlist.append(d5)
    dtwlist.append(e5)
    dtwlist.append(f5)
    dtwlist.append(h5)
    dtwlist.append(g5)
    #-------------5.part end---------



    closest=[]
    closest=min(dtwlist)#Returns the smallest of its parameters
    
    
    
    #----------------comparison and decision phase----------------------------
    if (closest==a1 or closest==a2 or closest==a3 or closest==a4 or closest==a5):
        result.append("a")
    elif (closest==b1 or closest==b2 or closest==b3 or closest==b4 or closest==b5):
        result.append("b")
    elif (closest==c1 or closest==c2 or closest==c3 or closest==c4 or closest==c5):
        result.append("c")
    elif (closest==d1 or closest==d2 or closest==d3 or closest==d4 or closest==d5):
        result.append("d")
    elif (closest==e1 or closest==e2 or closest==e3 or closest==e4 or closest==e5):
        result.append("e")
    elif (closest==f1 or closest==f2 or closest==f3 or closest==f4 or closest==f5):
        result.append("f")
    elif (closest==g1 or closest==g2 or closest==g3 or closest==g4 or closest==g5):
        result.append("g")
    elif (closest==h1 or closest==h2 or closest==h3 or closest==h4 or closest==h5):
        result.append("h")
    else:
        result.append("invalid letter")
    #------------------------End--------------------------------------------------
    
print(result)#printing result of decision
```
